### chinese dataset random baseline

```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, f1_score
```

```python
# read the csv datasets
train_df = pd.read_csv('train_zh_dataset.csv')
test_df = pd.read_csv('test_zh_dataset.csv')
```

```python
zh_test_label = test_df['label'].tolist()
```

```python
random_predictions = np.random.choice(np.unique(zh_test_label), size=len(zh_test_label))
```

```python
accuracy = accuracy_score(zh_test_label, random_predictions)
f1 = f1_score(zh_test_label, random_predictions, average='macro')
```

```python
accuracy, f1
```
